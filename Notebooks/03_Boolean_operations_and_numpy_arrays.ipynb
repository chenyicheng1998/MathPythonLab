## Boolean array indexing

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
b = x > 5
print('x = ', x)
print('b = ', b)
```

```python
x = np.arange(-5, 5)
y = 0.25*x**2
b = y > 2
print('y = ', y)
print('b = ', b)
```

## More examples of boolean array operations

```python
# Create an example array
x = np.arange(10)
x
```

```python
# Which elements are less than 4?
1.0*(x < 4)
```

```python
# Calculate x**2, when x < 4
x**2*(x < 4)
```

```python
# Calculate: 
# x, when x < 4
# x**2, when x >= 4
x*(x < 4) + x**2*(x >= 4)
```

```python
# Display x, when 3 < x < 8, otherwise 0
x*((3 < x) & (x < 8))
```

```python
# Display 5, when 3 < x < 8, otherwise 0
((3 < x) & (x < 8))*5
```

```python
# Define a step function
def step(x):
    return 1.0*(x>=0)
```

```python
# Calculate
# x/2, when 3 <= x < 6
# otherwise 0
(step(x-3) - step(x - 6))*x/2
```

```python
# Same results, but now calculated without step(x) function
((x >= 3) & (x < 6))*x/2
```

# Plotting boolean values

```python
fig, ax = plt.subplots(2, figsize = (10, 8))

plt.sca(ax[0])
plt.plot(x, y, 'o-', color = 'b', label = '$y = 0.25y^2$')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.sca(ax[1])
plt.plot(x, b, 'o-', color = 'r', label = '$y > 2$')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('b')
plt.legend()
plt.show()
```

```python
x = np.arange(-5, 5, 0.01)
y = 0.25*x**2
b = y > 2

fig1, ax = plt.subplots(3, figsize = (10, 9), 
                        sharey = True, sharex = True)
plt.sca(ax[0])
plt.plot(x, y)
plt.grid(True)

plt.sca(ax[1])
plt.plot(x, b, color = 'r')
plt.grid(True)

plt.sca(ax[2])
plt.plot(x, b*y)
plt.grid(True)
plt.xlabel('x');
```

### Piecewise continuous function

```python
x = np.arange(-5, 5, 0.01)
y1 = x**2 - 15
y2 = 3*x
y = (x < 0)*y1 + (x >= 0)*y2

fig1, ax = plt.subplots(3, figsize = (10, 9), 
                        sharey = True, sharex = True)

plt.sca(ax[0])
plt.plot(x, y1, label = 'y1')
plt.grid(True)
plt.legend(loc = 'lower right')

plt.sca(ax[1])
plt.plot(x, y2, label = 'y2')
plt.grid(True)
plt.legend(loc = 'lower right')

plt.sca(ax[2])
plt.plot(x, y, label = 'y')
plt.grid(True)
plt.legend(loc = 'lower right')
plt.xlabel('x');
```
