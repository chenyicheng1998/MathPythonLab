## Parts of a Figure

<img src="https://matplotlib.org/stable/_images/anatomy.png" alt="Parts of a figure" width="600">

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create example data, the 4th order polynomial function
x = np.linspace(0, 4)
y = 0.1*x**4 - 0.5*x**3 + 0.2*x**2 - 0.1*x + 4

fig, ax = plt.subplots(figsize = (10, 6))  # a figure with a single Axes

plt.plot(x, y, linestyle = '--', linewidth = 3, label = '4th order polynomial')
plt.grid(True)
plt.xlabel('x', fontsize = 14)
plt.ylabel('y', fontsize = 14)
plt.title('Graph of a function', fontsize = 20)
plt.legend()
plt.show()
```

See: [Controlling line properties](https://matplotlib.org/stable/tutorials/introductory/pyplot.html#controlling-line-properties)

## Working with multiple line plots

```python
fig, ax = plt.subplots()  # a figure with a single axes

x = np.linspace(0, 4)
y1 = x - 2
y2 = x**2 - 4*x

plt.plot(x, y1, label = '$y = x - 2$')
plt.plot(x, y2, label = '$y = x^2 - 4$')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Working with multiple axes 

```python
fig, ax = plt.subplots(2, figsize = (10, 8))  # Create a figure containing two axes

x = np.linspace(0, 4)
y1 = x - 2

plt.sca(ax[0])
plt.plot(x, y1, color = 'b', label = 'x - 2')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

y2 = x**2 - 4*x

plt.sca(ax[1])
plt.plot(x, y2, color = 'r', label = '$x^2 - 4x$')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Working with multiple figures

```python
fig1, ax1 = plt.subplots()  # Create a figure containing an axes

x = np.linspace(0, 4)
y1 = x - 2

plt.plot(x, y1, color = 'b', label = 'x - 2')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

y2 = x**2 - 4*x

fig2, ax2 = plt.subplots()  # Create another figure containing an axes

plt.plot(x, y2, color = 'r', label = '$x^2 - 4x$')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

More examples, see: [Basic usage of matplotlib](https://matplotlib.org/stable/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py)
