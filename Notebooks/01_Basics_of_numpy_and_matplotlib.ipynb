## Basic numpy examples

```python
import numpy as np
```

```python
# Create an array from list
x = np.array([0, 1, 2, 3, 4])
x
```

```python
# Another example
x = np.array([-2, -1, 0, 1, 2])
print(x)
```

```python
# Create an array using array range function
x = np.arange(5)
print(x)
```

```python
# Another example
x = np.arange(6)
print(x)
```

```python
# Third example
x = np.arange(-2, 3)
print(x)
```

```python
# Fourth example
x = np.arange(-2, 2.1, 0.5)
print(x)
```

```python
# Create an array using linspace
x = np.linspace(0, 5)
print(x)
```

```python
# Another example
x = np.linspace(0, 5, 6)
print(x)
```

```python
# Third example
x = np.linspace(-2, 2, 9)
print(x)
```

```python
# Array numerical computations
x = np.arange(6)
y = x**2
print(x)
print(y)
```

```python
# Second example
x = np.array([0, 1, 4, 9, 16, 25])
y = np.sqrt(x)
print(x)
print(y)
```

```python
# Third example, use linspace
x = np.linspace(-2, 2, 9)
y = 2*x + 1
print(x)
print(y)
```

## Draw a first plot - example
See: https://matplotlib.org/stable/users/getting_started/index.html#draw-a-first-plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 4, 0.05)
y = np.sin(x*np.pi)

fig, ax = plt.subplots(figsize=(6,4), constrained_layout=True)
ax.plot(x, y)
ax.set_xlabel('Time [s]')
ax.set_ylabel('Amplitude [V]')
ax.set_title('Sine wave')
fig.set_facecolor('lightsteelblue')
```

## Draw a first plot with pyplot

Adapted from the example above and modified to work with pyplot.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 4, 0.05)
y = np.sin(x*np.pi)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude [V]')
plt.title('Sine wave')
plt.xlim(0, 4)
plt.ylim(-1.1, 1.1)
plt.grid(True)
plt.axhline(0.0, color = 'red')
plt.axvline(2.0, color = 'red')
plt.show()
```
