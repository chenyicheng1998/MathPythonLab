## Exponential functions

There is a function for $y = e^x$

```python
x = np.arange(5)
y = np.exp(x)
print(y)
```

$y = e^{-x}$ can be calculated

```python
y = np.exp(-x)
print(y)
```

## $y = e^x$ and $y = e^{-x}$

```python
x = np.linspace(-2, 2)
y1 = np.exp(1.0*x)
y2 = np.exp(1.2*x)
y3 = np.exp(1.5*x)
plt.plot(x, y1, label = '$y_1 = e^{x}$')
plt.plot(x, y2, label = '$y_2 = e^{1.2x}$')
plt.plot(x, y3, label = '$y_2 = e^{1.5x}$')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Logarithm functions

$y = ln(x) \leftrightarrow x = e^y$

```python
x = np.array([1, 2, 5, 10, 20, 50])
y = np.log(x)
print(y)

x2 = np.exp(y)
print(x2)
```

$y = log(x) \leftrightarrow x = 10^y$

```python
x = np.array([1, 2, 5, 10, 20, 50])
y = np.log10(x)
print(y)

x2 = 10**y
print(x2)
```

Two-base logarithm $y = log_2(x) \leftrightarrow x = 2^y$

```python
x = np.array([1, 2, 4, 8, 16])
y = np.log2(x)
print(y)

x2 = 2**y
print(x2)
```

## semilog and loglog scales

Three functions in linear scales (plt.plot).

```python
x = np.array([1.5, 2.0, 5.0, 10.0, 20.0, 50.0, 80.0, 100.0])
y1 = 10*x + 1
y2 = 220.0*np.log(x)
y3 = np.exp(0.069*x)
plt.plot(x, y1, 'o-', label = '$y_1 = 10x + 1$')
plt.plot(x, y2, 's-', label = '$y_2 = 220.0 \cdot \log(x)$')
plt.plot(x, y3, 's-', label = '$y_3 = e^{0.069x}$')
plt.grid()
plt.legend()
plt.show()
```

### semilogx scale

```python
plt.semilogx(x, y1, 'o-', label = '$y_1 = 10x + 1$')
plt.semilogx(x, y2, 's-', label = '$y_2 = 220.0 \cdot \log(x)$')
plt.semilogx(x, y3, 's-', label = '$y_3 = e^{0.069x}$')
plt.grid()
plt.legend()
plt.show()
```

## semilogy scale

```python
plt.semilogy(x, y1, 'o-', label = '$y_1 = 10x + 1$')
plt.semilogy(x, y2, 's-', label = '$y_2 = 220.0 \cdot \log(x)$')
plt.semilogy(x, y3, 's-', label = '$y_3 = e^{0.069x}$')
plt.grid()
plt.legend()
plt.show()
```

## loglog scale

```python
plt.loglog(x, y1, 'o-', label = '$y_1 = 10x + 1$')
plt.loglog(x, y2, 's-', label = '$y_2 = 220.0 \cdot \log(x)$')
plt.loglog(x, y3, 's-', label = '$y_3 = e^{0.069x}$')
plt.grid()
plt.legend()
plt.show()
```

### What is the value of $e$?
$e = \sum_{n=0}^{\infty} \frac{1}{n!} = 1 + \frac{1}{1} + \frac{1}{1\cdot 2} + \frac{1}{1\cdot 2 \cdot 3} + ... \approx 2.718$

### How to write $x^2$ in label?
Use dollars '$$' around the string to write a LaTeX equation.
- Subscript x_0 -> $x_0$
- Superscript x^2 -> $x^2$
- Examples
    - $z = \sqrt{x^2 + y^2}$
    - $x - 2 \ge y$
- Learn more, try [Online latex editor](https://latexeditor.lagrida.com/)

```python
x = np.arange(10)
y = x**2
plt.plot(x,y, label = '$y = x^{2}$')
plt.legend()
plt.show()
```

### How do I calculate $e^{-t/RC}$?

```python
# Example
t = 0.02 # seconds
R = 2000 # ohms 
C = 10e-6 # farads, F
np.exp(-t/(R*C))
```
