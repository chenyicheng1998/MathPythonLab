## Problem 5
If $ A(a, x) = \left( \begin{matrix}
a + x & a - x & a - x \\
a - x & a + x & a - x \\
a - x & a - x & a + x
\end{matrix} \right)$ and $a = 3.0$, find $x > 0$, so that $det(A(a, x)) = 0$. 

How are $a$ and $x$ related? When you have found the relationship, try another value for $a$.

Tip: Make a code that 
- uses linearly spaced values for $x \in [0, 10]$, for example. Use `numpy.linspace`.
- calculates the $f(x) = det(A(a, x))$ for all values of $x$ and then 
- makes a graphical plot of $f(x)$ versus $x$. 
- Finally, from the graph, find when $f(x) = 0$.

```python
# Solution for problem 5
def matrix_A(a, x):
    return np.array([[a + x, a - x, a - x],
                     [a - x, a + x, a - x],
                     [a - x, a - x, a + x]])

x_values = np.linspace(0, 15, 1000)

a = 3.0

det_values = [np.linalg.det(matrix_A(a, x)) for x in x_values]

plt.plot(x_values, det_values)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('f(x) vs x, when (a = 3)')
plt.grid(True)
plt.show()


a_values = [1, 2, 3, 4]

for a in a_values:
    det_values = [np.linalg.det(matrix_A(a, x)) for x in x_values]
    plt.plot(x_values, det_values, label=f'a = {a}')
    
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('f(x) vs x, when (a = 1 or 2 or 3 or 4)')
plt.grid(True)
plt.legend()
plt.show()

```
