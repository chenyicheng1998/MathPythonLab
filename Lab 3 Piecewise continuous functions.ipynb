## Problem 1

The unit step function

$ u(t) = \begin{cases}
1 & t \geq 0 \\
0 & t < 0 
\end{cases}$

can be coded using numerical Python as given below.

Sketch a graph of the unit step function $u(t)$ and annotate the graph.
- Title: "Unit step function"
- x-label: "Time (s)"
- y-label: "Amplitude"
- Add grid-lines

Notice that the step-function and the values for *t* are already given, and you can use them to calculate the values of *f(t)*.

```python
t = np.arange(-1, 5, 0.01)
u = 1.0*(t >= 0)

plt.plot(t, u)
plt.grid(True)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Unit step function')
plt.show()
```

## Problem 2

A delayed unit step function is defined as

$ u(t-t_0) = \begin{cases}
1 & t \geq t_0 \\
0 & t < t_0
\end{cases}$

That can be coded just by replacing the value 0 to t0.

Sketch a graph of delayed unit step function $u(t-3)$. Annotate the graph.
- Title: "Delayed unit step function"
- x-label: "Time (s)"
- y-label: "Amplitude"
- Add grid-lines

Notice that the values for *t* are already given, and you can use them to calculate the values of *u(t - 3)*.

```python
t = np.arange(-1, 5, 0.01)
u = 1.0*(t >= 3)

plt.plot(t, u)
plt.grid(True)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Delayed unit step function')
plt.show()
```

## Problem 3

[Rectangular pulse](https://en.wikipedia.org/wiki/Rectangular_function) can be defined as

$ p(t, t_0, t_1) = \begin{cases}
1 & t_0 \leq t < t_1 \\
0 & \text{otherwise}
\end{cases}$

We can implement it using two delayed step functions and subtracting them, like $p(t, t_0, t_1) = u(t-t_0) - u(t-t_1)$. 

Sketch a graph of a rectangular pulse function $p = u(t-1) - u(t-3)$. Annotate the graph.
- Title: "Rectangular pulse function"
- x-label: "Time (s)"
- y-label: "Amplitude"
- Add grid-lines


```python
t = np.arange(-1, 5, 0.01)
p = 1.0*(t >= 1) - 1.0*(t >= 3)

plt.plot(t, p)
plt.grid(True)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Rectangular pulse function')
plt.show()
```

## Problem 4

By multiplying the rectangular pulse with another function, we can create pulse functions with arbitrary shape.

Sketch a graph of a polynomial pulse $f = (-t^2 + 4t - 3) \cdot \left( u(t-1) - u(t-3) \right)$. 

Annotate the graph.
- Title: "Polynomial pulse function"
- x-label: "Time (s)"
- y-label: "Amplitude"
- Add grid-lines


```python
t = np.arange(-1, 5, 0.01)
f = (-t**2+4*t-3)*(1.0*(t >= 1) - 1.0*(t >= 3))

plt.plot(t, f)
plt.grid(True)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Polynomial pulse function')
plt.show()
```

## Problem 5

Calculate the values for a [triangular function](https://en.wikipedia.org/wiki/Triangular_function)

$$f = \begin{cases}
t/2 & 0 \leq t < 2 \\
-t/2 + 2 & 2 \leq t < 4 \\
0 & \text{otherwise}
\end{cases}$$

Sketch a graph of the function and annotate the graph.
- Title: "Triangular pulse function"
- x-label: "Time (s)"
- y-label: "Amplitude"
- Add grid-lines

Tip: First, calculate the parts $f_1(t) = t/2$ (when $0 \leq t < 2$) and $f_2(t) = -t/2 + 2$ (when $2 \leq t < 4$), and then sum them together.

```python
t = np.arange(-1, 5, 0.01)
f = t/2 * (1.0*(t >= 0) - 1.0*(t > 2)) + (-t/2+2) * (1.0*(t >= 2) - 1.0*(t > 4))

plt.plot(t, f)
plt.grid(True)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Triangular pulse function')
plt.show()
```
